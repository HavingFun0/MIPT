# Задача 7.18


Ищем решение разностного уравнения в виде $y_k \propto q^k: q^2 - (2-\lambda h^2)q + 1 = 0, q_1q_2 = 1$, следовательно общее решение выглядит, как $y_k = \alpha q_1^k + \beta q_1^{-k}$. Запишем граничные условия:
$$ \alpha + \beta = \alpha q + \frac{\beta}{q} \rightarrow \alpha = \frac{\beta}{q} $$
$$ \beta q^{n-1} + \beta q^{-n} = 0 \rightarrow q^{2n-1} = 1 \rightarrow q = e^{i\frac{2\pi k}{2n-1}} $$


$y_k = \beta(q^{k-1} + q^{-k})$

Вырожденный случай: $q_1 = q_2 = -1: y_k = \alpha (-1)^k + \beta k (-1)^k$. Подставим граничные условия: 
\begin{equation*}
\begin{cases}
\alpha = -\alpha - \beta \\
\alpha (-1)^N + \beta N (-1)^N = 0
\end{cases}
\end{equation*}

Видно, что матрица уравнений на коэффициенты невырождена $\Rightarrow \alpha = \beta = 0$ 

Найдём сами собственные значения: $2-\lambda_j h^2 = 2\cos(\frac{2\pi j}{2N-1})$

$$\lambda_j = \frac{4}{h^2}\sin^2(\frac{\pi j}{2N-1}), j = \overline{0, N-1}$$

# Задача 7.20



Начало похожее, но граничные условия теперь выглядят так:
$$ \alpha + \beta = \alpha q + \frac{\beta}{q} \rightarrow \alpha = \frac{\alpha}{q} $$
$$ \beta (q^{N-1} + q^{-N}) = \beta (q^{N-2} + q^{-N+1}) \rightarrow q^{2N-1}\frac{q-1}{q} = q-1 \rightarrow q^{2N-2} = 1$$

Теперь, если проверять вырожденный случай, получим: 
\begin{equation*}
\begin{cases}
\alpha = -\alpha - \beta \\
\alpha (-1)^N + \beta N (-1)^N =  \alpha (-1)^{N-1} + \beta (N-1) (-1)^{N-1}
\end{cases}
\end{equation*}
и, опять этот случай не реализуется

Теперь найдём собственные числа: $2-\lambda_j h^2 = 2\cos(\frac{2\pi j}{2(N-1)})$

$$\lambda_j = \frac{4}{h^2}\sin^2(\frac{\pi j}{2(N-1)}), j = \overline{0, N-1}$$

# Задача 7.9


Характеристическое уравнение у этой разностной схемы: $ 7q^5 - 2q^4 - (8\alpha + 5)q^3 + (8\alpha + 3)q^2 - 2q - 1 = 0 $. Любое её решение будет ограниченным при $k\to\infty$ тогда и только тогда, когда все корни хар. ур-ия по модулю меньше 1.
Корень 1 угадаем сразу и перепишем это, как $(q-1)(7q^4 + 5q^3 - 8\alpha q^2 + 3q + 1) = 0$. Воспользуемся теоремой Шура-Кона: модуль семи больше модуля одного, проверим что там с уравнением 
$$ \frac{48q^4 + 32q^3 - 48\alpha q^2 + 16q}{q}  = 0$$
$$ 3q^3 + 2q^2 - 3\alpha q + 1 = 0 $$
Хорошая теорема, можно повторить её применение
$$ 8q^2 + (6-3\alpha)q - (9\alpha + 2) = 0 $$
Тут потребуем $8 > |9\alpha + 2|$ и понизим степень ещё раз:
$$ (64 - (9\alpha + 2)^2)q + (8(6-3\alpha) + (9\alpha + 2)(6 - 3\alpha)) = 0 \Leftrightarrow q = -\frac{(6-3\alpha)(9\alpha + 10)}{(6-9\alpha)(9\alpha + 10)} $$
Итого: $9\alpha \in (-10;6)$ и $|2-\alpha| < |2 - 3\alpha|$. Последнее верно при $\alpha \in (-\infty; 0)\cup(1;+\infty)$

Ответ: $\alpha \in (-\frac{10}{9}; 0)$

# Задача 7.12-4


Решим характеристическое уравнение: 

\begin{equation*}
\begin{vmatrix}
 10 - q & -6\\
 18 & -11 - q
\end{vmatrix}
 = (q+11)(q-10) + 108 = 0
\end{equation*}

Решения: $q = 1 \vee q = -2$, соответствующее им собственные вектора $(2;3)$, и $(1;2)$, таким образом общее решение однородной системы это

\begin{equation*}
\left(
    \begin{array}{c}
    x_n\\
    y_n
    \end{array}
\right) = \alpha \left(
    \begin{array}{c}
    2\\
    3
    \end{array}
\right) + \beta \left(
    \begin{array}{c}
    1\\
    2
    \end{array}
\right)(-2)^n
\end{equation*}

Решение же неоднородного уравнения будем угадывать в виде 
\begin{equation*}
\left(
\begin{array}{c}
x^*_n \\
y^*_n
\end{array}
\right) = \left(
\begin{array}{c}
    \alpha(n+a)2^n \\
    \beta(n+b)2^{n}
\end{array}
\right)
\end{equation*}

Подставляем:

\begin{equation*}
\left(
\begin{array}{c}
\alpha(n+a+1)2^{n+1} \\
\beta(n+b+1)2^{n+1}
\end{array}
\right) = \left(
\begin{array}{c}
    10n\alpha 2^n - 6n\beta 2^n - 2\cdot2^n\\
    18n\alpha 2^n - 11n\beta 2^n - 5\cdot2^n
\end{array}
\right)
\end{equation*}

Сократим слагаемые с $n2^N$:

\begin{equation*}
\begin{cases}
    9\alpha - 6\beta = 0 \\
    18\alpha - 12\beta = 0
\end{cases}
\end{equation*}

Отсюда имеем, что $\alpha = 2k, \beta = 3k$. Теперь ищем $k$ и подгоняем $a$ и $b$ чтобы сократить вообще всё.


\begin{equation*}
\begin{cases}
    2ak = -1\\
    3bk = -5/2
\end{cases}
\end{equation*}

Мне вот нравится $k = -\frac{1}{6}$, тогда $a = 3, b = 5$

Итого, частное решение:

\begin{equation*}
\left(
\begin{array}{c}
x^*_n \\
y^*_n
\end{array}
\right) = \left(
\begin{array}{c}
    \frac{1}{3}(n+3)2^n \\
    \frac{1}{2}(n+5)2^{n}
\end{array}
\right)
\end{equation*}

 2 - q & 3\\

\end{equation*}
