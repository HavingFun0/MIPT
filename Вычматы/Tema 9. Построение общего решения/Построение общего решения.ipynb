## Задача 2(7.18)


Напишем характеристическое уравнение
$$ q^2 + (\lambda h^2 - 2) q + 1 = 0 \Rightarrow q_1 = \frac{1}{q_2}$$
Теперь запишем граничные условия, с учетом $y_n = a q_1^n + \dfrac{1}{q_1^n}$
$$ \begin{cases}  
a + b = a q_1 + \dfrac{b}{q_1} \\
a q_1^N + b q_1^{-N} = 0
\end{cases} 
\Rightarrow \begin{cases}  
a(1 - q_1^{2N}) = a (q_1 - q_1^{2N-1}) \\
b = -a q_1^{2N}
\end{cases} \Rightarrow  \begin{cases}  
q_1^{2N-1}(1 - q_1) = q_1 - 1 \\
b = -a q_1^{2N}
\end{cases} \Rightarrow 
\begin{cases}  
q_1^{2N-1} = - 1 \\
b = -a q_1^{2N}
\end{cases}  \Rightarrow $$
$$\begin{cases}  
q_1 = \exp{\dfrac{i \pi (2k-1)}{2N-1}}, k = 1,...,N-1 \\
\\
b = - a  \exp{\dfrac{i \pi (2k-1)2N}{2N-1}} = a \exp{\dfrac{i \pi (2k-1)}{2N-1}} 
\end{cases}  $$
k выбраны таким образом чтобы не включать тривиальных решений и чтобы решения не повторялись, учитывая $q_2=\frac{1}{q1}$. \\
Тогда получаем общий вид претендента на решение, удовлетворяющего граничным условиям
$$y_n = a (\exp{\dfrac{i \pi (2k-1)n}{2N-1}} + \exp{\dfrac{i \pi (2k-1)(1-n)}{2N-1}}) 
= \alpha \cos{\dfrac{\pi (2k-1)(n-\frac{1}{2})}{2N-1}} = \alpha \cos{\theta_k (n-\frac{1}{2})} $$
Подставим и получим
$$\lambda_k = \dfrac{-N^2 (\cos{\theta_k (n+\frac{1}{2})} - 2\cos{\theta_k (n-\frac{1}{2})}+\cos{\theta_k (n-\frac{3}{2})})}{\cos{\theta_k (n-\frac{1}{2})}} = 2 N^2(1-\cos{\theta_k})$$
Таким образом все возможные собственные значения
$$\lambda_k = 4 N^2 \sin^2{\dfrac{\pi(2k-1)}{2(2N-1)}}, k=1,...,N-1$$


## Задача 3(7.20)


Напишем характеристическое уравнение
$$ q^2 - (\lambda h^2 + 2) q + 1 = 0 \Rightarrow q_1 = \frac{1}{q_2}$$
Теперь запишем граничные условия, с учетом $y_n = a q_1^n + \dfrac{1}{q_1^n}$

$$ \begin{cases}  
a q_1^N + b q_1^{-N} = a q_1^{N-1} + b q_1^{1-N} \\
a + b = 0 
\end{cases} 
\Rightarrow \begin{cases}  
a(q_1^{2N}-1) = a (q_1^{2N-1}-q_1) \\
a=-b
\end{cases} \Rightarrow   \begin{cases}  
q_1^{2N-1}(q_1-1) = 1-q_1 \\
b = -a 
\end{cases} \Rightarrow \begin{cases}  
q_1^{2N-1} = - 1 \\
b = -a 
\end{cases}$$

$$
\Rightarrow
\begin{cases}  
q_1 = \exp{\dfrac{i \pi (2k-1)}{2N-1}}, k = 1,...,N-1 \\
\\
b = - a
\end{cases}  $$
Таким образом кандидат на решение, удовлетворяющий граничным условиям
$$y_n = \alpha \sin{ \dfrac{\pi(2k-1)n}{2N-1}} = \alpha \sin{\theta_k n} $$
Получаем собственные числа
$$\lambda_k = - \dfrac{\sin{\theta_{k} (n+1)} - 2\sin{\theta_k n} + \sin{\theta_{k} (n-1)}}{h^2 \sin{\theta_k n}} = \dfrac{2(1-\cos{\theta_k})}{h^2}$$
Таким образом все возможные собственные значения
$$\lambda_k = 2N^2\left(1-\cos{\dfrac{\pi(2k-1)}{2N-1}}\right), k=1,...,N-1$$

## Задача 4(7.9)


Характеристическое уравнение у этой разностной схемы: $ 7q^5 - 2q^4 - (8\alpha + 5)q^3 + (8\alpha + 3)q^2 - 2q - 1 = 0 $. Любое её решение будет ограниченным при $k\to\infty$ тогда и только тогда, когда все корни хар. ур-ия по модулю меньше 1.
Корень 1 угадаем сразу и перепишем это, как $(q-1)(7q^4 + 5q^3 - 8\alpha q^2 + 3q + 1) = 0$. 

Воспользуемся теоремой Шура-Кона: модуль семи больше модуля одного, проверим что там с уравнением 
$$ \frac{48q^4 + 32q^3 - 48\alpha q^2 + 16q}{q}  = 0$$
$$ 3q^3 + 2q^2 - 3\alpha q + 1 = 0 $$
Снова применим теорему

$$ 8q^2 + (6-3\alpha)q - (9\alpha + 2) = 0 $$

Потребуем $8 > |9\alpha + 2|$ и понизим степень ещё раз:
$$ (64 - (9\alpha + 2)^2)q + (8(6-3\alpha) + (9\alpha + 2)(6 - 3\alpha)) = 0 \Leftrightarrow q = -\frac{(6-3\alpha)(9\alpha + 10)}{(6-9\alpha)(9\alpha + 10)} $$
Итого: $9\alpha \in (-10;6)$ и $|2-\alpha| < |2 - 3\alpha|$. Последнее верно при $\alpha \in (-\infty; 0)\cup(1;+\infty)$

Ответ: $\alpha \in (-\frac{10}{9}; 0)$

## Задача 5(7.12-4)



Решим характеристическое уравнение: 

\begin{equation*}
\begin{vmatrix}
 10 - q & -6\\
 18 & -11 - q
\end{vmatrix}
 = (q+11)(q-10) + 108 = 0
\end{equation*}

Решения: $q = 1, q = -2$, соответствующее им собственные вектора $(2;3)$, и $(1;2)$, таким образом общее решение однородной системы это



\begin{equation*}
\left(
    \begin{array}{c}
    x_n\\
    y_n
    \end{array}
\right) = \alpha \left(
    \begin{array}{c}
    2\\
    3
    \end{array}
\right) + \beta \left(
    \begin{array}{c}
    1\\
    2
    \end{array}
\right)(-2)^n
\end{equation*}

Решение же неоднородного уравнения будем искать в виде 
\begin{equation*}
\left(
\begin{array}{c}
x^*_n \\
y^*_n
\end{array}
\right) = \left(
\begin{array}{c}
    \alpha(n+a)2^n \\
    \beta(n+b)2^{n}
\end{array}
\right)
\end{equation*}

Подставляем:

\begin{equation*}
\left(
\begin{array}{c}
\alpha(n+a+1)2^{n+1} \\
\beta(n+b+1)2^{n+1}
\end{array}
\right) = \left(
\begin{array}{c}
    10n\alpha 2^n - 6n\beta 2^n - 2\cdot2^n\\
    18n\alpha 2^n - 11n\beta 2^n - 5\cdot2^n
\end{array}
\right)
\end{equation*}

Слагаемые с $n2^N$:

\begin{equation*}
\begin{cases}
    9\alpha - 6\beta = 0 \\
    18\alpha - 12\beta = 0
\end{cases}
\end{equation*}

Отсюда имеем, что $\alpha = 2k, \beta = 3k$. Теперь ищем $k$, $a$ и $b$.

\begin{equation*}
\begin{cases}
    2ak = -1\\
    3bk = -5/2
\end{cases}
\end{equation*}


Возьмем $k = -\frac{1}{6}$, тогда $a = 3, b = 5$

Итого, частное решение:

\begin{equation*}
\left(
\begin{array}{c}
x^*_n \\
y^*_n
\end{array}
\right) = \left(
\begin{array}{c}
    \frac{1}{3}(n+3)2^n \\
    \frac{1}{2}(n+5)2^{n}
\end{array}
\right)
\end{equation*}
