$$y_{n+1}-(2+(10 + x_n)h^2)y_n+y_{n-1}= f_n h^2$$

1) $y'' + (10+x)y=0, y(0)=1, y'(0)=0$

2) $y'' + (10+x)y=0, y(0)=0, y'(0)=1$

3) $y'' + (10+x)y=x e^{-x}, y(0)=0, y'(0)=0$


**Ответы на вопросы:**
3 задачи Коши нужно для определения общего решения, ведь для уравнения первого порядка справедливо, что $y(x) = C_1 y_1(x) + C_2 y_2(x) + y_0(x)$, где $y_0(x)$ --- какое-либо решение неоднородного уравнения $y'' + q(x) y' - p(x)y = f(x)$, а $y_1(x)$ и $y_2(x)$ --- два любые линейно независимые решения однородного уравнения $y'' + q(x) y' - p(x)y = 0$. Эти задачи могут быть решены, методом прогонки. Для определения значений сеточной функции получается СЛАУ с трехдиагональной матрицей 

\begin{equation*}
      \textbf{A}' = \left( \begin{array}{ccccc}
      -b_1 & c_1 & 0 & \dots & 0\\
       a_2 & -b_2 & c_2 & \dots & 0 \\
       0 & a_3 & -b_3 & \dots & 0\\
       \dots & \dots & \dots & \dots & \dots\\
       0 & 0 & 0 &  \dots & -b_{N-1}
       \end{array}\right),
\end{equation*}
где введены обозначения

\begin{equation*}
    a_n = 1, \ c_n = 1, \ b_n = 2 + (10 + nh) h^2, \ n=1,\dots,N-1;
\end{equation*}

Для устойчивости метода прогонки достаточно диагонального преобладания матрицы, но тут можно наблюдать даже строгое диагональное преобладание, т.к.
\begin{equation*}
    \left| b_n \right| > \left| a_n \right| + \left| c_n \right| \ \forall n
\end{equation*}
\begin{equation*}
    \left| 2 + (10 + nh) h^2 \right| > \left| 2 \right| \ \forall n.

## Задача 10(9.1а)


 Будем искать решение для неоднородного уравнения с условиями $y_0 = 0, y'(10) \approx \frac{y_N - y_{N-1}}{h} = 0 \Rightarrow y_N = y_{N-1}$. Условия для решений однородного:




Данное уравнение жёсткое. Условия для решений однородного:

$$
\begin{cases}
y1: \, y_0 = 0, y_N = y_{N-1} + h\\
y2: \, y_0 = 1, y_N = y_{N-1}
\end{cases}
$$


Будем решать методом прогонки: 


$$\begin{cases}
y' = v\\
v' = (10+x^2)y + xe^{-x}
\end{cases}
$$

$y(x) = \alpha(x)v(x) + \beta(x), y' = \alpha'v + \alpha v' + \beta'$. Подставляем $y'$ и $v'$: $$ v = \alpha'v + \alpha^2(10+x^2)v + \alpha\beta(10+x^2) + \alpha xe^{-x} + \beta' $$

$$
\begin{cases}
    \alpha' = 1 - \alpha^2(10+x^2) \\
    \alpha\beta(10+x^2) + \alpha xe^{-x} + \beta' = 0
\end{cases}
$$
Найдём альфу:



```python
import numpy as np
import matplotlib.pyplot as plt
```

**Неоднородное с нулевыми краевыми условиями.**

```python
h = 0.01
X = np.linspace(0,10, int(10/h+1))
al = np.zeros(len(X))
al[0] = 0 #from y(0) = 0

for i in range(1, len(X)):
    al[i] = al[i-1] + h*(1 - (al[i-1]**2)*(10 + X[i - 1]**2))
    
plt.plot(X, al)

plt.grid()
plt.title('alpha')
plt.show()
```

```python
be = np.zeros(len(X))
be[0] = 0
for i in range(1, len(X)):
    x = X[i - 1]
    be[i] = be[i-1] - h*(al[i-1]*x*np.exp(-x) + al[i-1]*be[i-1]*(10 + x**2))
    
plt.plot(X, be)

plt.title('beta')
plt.grid()
plt.show()
```

Диффур на функцию $v$: $v' = \alpha(10+x^2)v + \beta(10+x^2) + xe^{-x}, v(10) = 0$.


Соотношение для частного решения: $v_{n-1} = v_n - h(\alpha_n(10+x_n^2)v_n + \beta_n(10+x^2_n) + x_ne^{-x_n} )$

```python
v = np.zeros(len(X))
v[len(X)-1] = 0
for i in range(len(X)-1, 0, -1):
    x = X[i]
    v[i-1] = v[i] - h*(al[i]*(10 + x**2)*v[i] + be[i]*(10 + x**2) + x*np.exp(-x) )
    
    
plt.plot(X, v)
plt.title('v')

plt.grid()
plt.show()
```

Окончательно

Решение неоднородного с краевыми условиями:

```python
y0 = al*v + be
plt.plot(X, y0)
plt.title('Решение неоднородного с нулевыми краевыми условиями')
plt.grid()
plt.show()
```

**Однородное первое**

```python
# Пересчитаем \beta без правой части и v

be1 = np.zeros(len(X))
be1[0] = 0
for i in range(1, len(X)):
    x = X[i - 1]
    be1[i] = be1[i-1] - h*(al[i-1]*be1[i-1]*(10 + x**2))
    
v1 = np.zeros(len(X))
v1[len(X)-1] = 1 # new boundary condition v(10) = 1
for i in range(len(X)-1, 0, -1):
    x = X[i]
    v1[i-1] = v1[i] - h*(al[i]*(10 + x**2)*v1[i] + be1[i]*(10 + x**2) )
```

```python
y1 = al*v1 + be1
plt.plot(X, y1)
plt.title('Решение однородного с одними краевыми условиями')
plt.grid()
plt.show()
```

**Второе однородное:**

```python
be2 = np.zeros(len(X))
be2[0] = 1
for i in range(1, len(X)):
    x = X[i - 1]
    be2[i] = be2[i-1] - h*(al[i-1]*be2[i-1]*(10 + x**2))
    
v2 = np.zeros(len(X))
v2[len(X)-1] = 0 # new boundary condition v(10) = 0
for i in range(len(X)-1, 0, -1):
    x = X[i]
    v2[i-1] = v2[i] - h*(al[i]*(10 + x**2)*v2[i] + be2[i]*(10 + x**2) )
```

```python
y2 = al*v2 + be2
plt.plot(X, y2)
plt.title('Решение однородного с другими краевыми условиями')
plt.grid()
plt.show()
```

## Задача 11(9.3а)

Разностная схема для ЗК с условием $y'(0)=a$

$\displaystyle \frac{y_{n+1}-2y_{n}+y_{n-1}}{h^2}=x_n \sqrt{y_n}$

$y_0 = 0$

Очевидно, что порядок аппроксимации -- $\omicron(dx^2)$. 

Реализуем неявную схему Эйлера

```python
def solve(a, N = 500):
    h = 1/N
    x, y, u = np.linspace(0, 1, N), np.linspace(0, 1, N), np.linspace(0, 1, N)

    y[0] = 0
    u[0] = a

    for i in range(2, N):
        y[i] = x[i-1]*np.sqrt(y[i-1])*(h**2)- y[i-2] + 2*y[i-1]

    for i in range(1,N):
        D = (2*u[i-1]+h*h*h*i*h*i*h)**2 - 4*(u[i-1]*u[i-1]-y[i-1]*h*h*i*h*i*h)
        u[i] = ((2*u[i-1]+h*h*h*i*h*i*h)+np.sqrt(D))/2
        y[i] = y[i-1]+u[i]*h

    return x, y
```

Решим методом деления отрезка пополам со стартовым отрезком $[1;2]$ c точностью $\varepsilon = 10^{-6}$

```python
a = 1
b = 2
c = (a + b) / 2
eps = 1e-6

while abs(solve(c)[1][-1] - 2) > eps:
    if (solve(c)[1][-1] - 2 > 0 ):
        b = c
    else:
        a = c
    c = (a + b) / 2

a_real = (a + b) / 2
print(a_real)
```

```python
x,y = solve(a_real)
```

```python
plt.plot(x ,y, label='График численного решения')

plt.title('Решения')
plt.ylabel('y')
plt.xlabel('x')
plt.grid()
plt.legend()
plt.show()
```
