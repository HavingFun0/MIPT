```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def progonka(y0,dy1,N,odnorodnost):
    P = np.linspace(0,10,N)
    Q = np.linspace(0,10,N)
    y = np.linspace(0,10,N)
    h = 10/N
    P[0] = 0
    Q[0] = y0
    #прямой ход
    for i in range(1,N-1):
        bi = 2+(10 + i*h)*h*h
        if odnorodnost:
            di=0
        else:
            di = h*h*i*h*np.exp(-i*h)
        P[i] = 1/(bi - P[i-1])
        Q[i] = (Q[i-1]-di)/(bi-P[i-1])
    Q[N-1] = (-Q[N-2] - h*dy1)/(-1+P[N-2])
    y[N-1] = Q[N-1]
    #обратный ход
    for i in range(N-2,0,-1):
        y[i] = P[i]*y[i+1]+Q[i]
    return y
```

```python
N = 10000
y = progonka(0,0,N,0)
x = np.linspace(0,10,N)
y = progonka(1,0,N,1)
plt.plot(x[1:],y[1:], label = "однородное1")
plt.grid()
plt.legend()
plt.show()
y = progonka(0,1,N,1)
plt.plot(x,y, label = "однородное2")
plt.grid()
plt.legend()
plt.show()
y = progonka(0,0,N,0)
plt.plot(x,y, label = "неоднородное")
plt.grid()
plt.legend()
plt.show()
```

```python
# def Cauchy(y0,dy0,N):
#     #максимально тупо перепишем оператор 2 и 1 производных
#     y = np.linspace(0,1,N)
#     h = 1/N
#     y[0] = y0
#     y[1] = h*dy0 + y0
#     for i in range(2,N):
#         y[i] = h*h*np.sqrt(y[i-1])*i*h + 2*y[i-1] - y[i-2]
#     return y
def Cauchy(y0,dy0,N):
    #реализуем неявный метод эйлера
    y = np.linspace(0,1,N)
    u = np.linspace(0,1,N)
    h = 1/N
    y[0] = y0
    u[0] = dy0
    for i in range(1,N):
        D = (2*u[i-1]+h*h*h*i*h*i*h)**2 - 4*(u[i-1]*u[i-1]-y[i-1]*h*h*i*h*i*h)
        u[i] = ((2*u[i-1]+h*h*h*i*h*i*h)+np.sqrt(D))/2
        y[i] = y[i-1]+u[i]*h
    return y
```

```python
#метод дихотомии с прошлого семака,
# мы ленивые поэтому он находит минимум функции (y(1)-2)^2
def f(a):
    N=100
    y = Cauchy(0,a,N)
    return (y[N-1]-2)*(y[N-1]-2)
def step(a,b,delta,f):
    u1 = (a+b-delta)/2
    u2 = (a+b+delta)/2
    if f(u1)>f(u2):
        return u1,b
    else:
        return a,u2
def dihot(a,b,eps,delta,f):
    while (b-a)>eps:
        a,b = step(a,b,delta,f)
    return((a+b)/2)
```

```python
delta = 0.0001
eps = 0.001
a = 0
b = 100
alpha = dihot(a,b,eps,delta,f)
print(alpha)
```

```python
x = np.linspace(0,1,1000) 
y = Cauchy(0,alpha,1000)
plt.plot(x,y, label = "решение")
plt.grid()
plt.legend()
plt.show()
```
