## Задача 2(7.1е)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t):
    return (t-5)*np.exp(t)
```

```python
a, b = 3, 6
eps = 0.001

while abs(b-a) >= eps:
    delta = (b-a)/100
    if f((a+b)/2 + delta) > f((a+b)/2 - delta):
        b = (a+b)/2 + delta
    else:
        a = (a+b)/2 - delta
        
print((a+b)/2)
```

```python
x = np.linspace(2,5,40)
plt.plot(x, f(x))
plt.vlines((a+b)/2, -60, -50, colors = 'red', linestyles='dashed')
plt.grid()
plt.show()
```

## Задача 3(7.6б)

```python
def f(x,y):
    return x**3 + 8*y**3 - 6*x*y + 1

def gradf(x,y):
    return np.array([3*x**2 - 6*y, 24*y**2 - 6*x])


eps = 0.001
```

Попробуем осуществить покоординатный спуск.

```python
def one_dim_min(a, b, f):
    while abs(b-a) >= eps:
        delta = (b-a)/100
        if f((a+b)/2 + delta) > f((a+b)/2 - delta):
            b = (a+b)/2 + delta
        else:
            a = (a+b)/2 - delta
    return (a+b)/2

ax, bx = -0.2, 1.2
ay, by = -0.2, 0.7
```

```python
u = np.array([1.1, 0.6])
i = 0
while np.linalg.norm(gradf(u[0], u[1])) >= 10*eps:
    x = one_dim_min(ax, bx, lambda x: f(x, u[1]))
    u[0] = x
    y = one_dim_min(ay, by, lambda x: f(u[0], x))
    u[1] = y
    i += 1
    if i > 1000:
        print('нет экстремума')
        break

print(f(u[0], u[1]), u)
```

Метод показывает, что экстремум в точке (1, 0.5). Во-первых в ней градиент обнуляется, во-вторых проверим это графически.

```python
x = np.linspace(-1.5, 1.5, 20)
plt.plot(x, f(x,1/2))

plt.grid()
plt.xlabel('x')
plt.ylabel('f(x, 1/2)')
plt.show()
```

```python
y = np.linspace(0.0, 0.8, 20)
plt.plot(y, f(1,y))

plt.grid()
plt.xlabel('y')
plt.ylabel('f(1, y)')
plt.show()
```
