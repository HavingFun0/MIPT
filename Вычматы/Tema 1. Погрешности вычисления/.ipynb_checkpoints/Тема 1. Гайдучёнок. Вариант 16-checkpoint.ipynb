## Задача 8.32

### Решение

Найдём полную погрешность вычисления производной. Она складывается из погрешности схемы и погрешности вычисления 

1) Найдём погрешность вычисления первой. Будем считать, что $g_{выч} = g(1+\epsilon)$ $\Rightarrow$

$r_{выч} \leq |\epsilon| \dfrac{g_{i-2} + 8 g_{i-1} +  8 g_{i+1} + g_{i+2}}{12 \tau} = \dfrac{3\epsilon M}{2 \tau}$

Где $M = \max(g_{i-2}, g_{i-1}, g_{i+2}, g_{i+1})$ по всем i.

2) Найдём погрешность метода. Формула имеет четвертый порядок точности $\Rightarrow$ с помощью разложения в ряд Тейлора по $\tau$ найдем погрешность схемы вычисления производной.

Учтя, что $u^{(5)}[t] \leq M_5$ получаем $r_{м} \leq  \dfrac{M_5 \tau^4}{30}$ 

Собирая погрешности вместе получим 

$r \leq r_{выч} + r_{м} = \dfrac{M_5 \tau^4}{30} + \dfrac{3\epsilon M}{2 \tau}$

Ищя экстремум этого значения(в данном случае он один и это, очевидно, минимум) равен:

**Ответ:**$\tau =\sqrt[5]{ \dfrac{45 \epsilon M}{2 M_5}}$

Как видно для положительных значений всё вычисляется с точность до $\epsilon_{маш}$, которое в одинарное точности порядка $10^{-7}$.

Для отрицательных же всё ломается, т.к. числа становятся сначала по модулю очень большими, а потом term становится соизмеримыми с погрешностями округления и программа заканчивает расчёт на полученном значении.

Для целых отрицательных значений экспоненты можно зная число значение $e^x$ и беря просто обратное к этому числу, т.е. следуя формуле $e^{-x} = \dfrac{1}{e^{x}}$.

Попробуем оценить погрешность вычисления экспоненты этим способом считая, что числа имеют погрешности, а операции -- нет.

Из индукции легко можно показать, что погрешность $x^k$ равна  
$$ x^k = x(1 + k \delta)$$

Подставляя это в сумму

$$\sum_{n=0}^N \dfrac{x^n (1+n\delta)}{n!} = \sum_{n=0}^N \dfrac{x^n}{n!} + \sum_{n=0}^{N-1} \dfrac{|x|^{n+1} \delta}{n!}= \sum_{n=0}^N \dfrac{x^n}{n!}  + \sum_{n=0}^N \dfrac{|x|^{n+1}}{n!} - \dfrac{x^{N+1}}{N!} \delta$$

Причем известно, что последнее слагаемое при делении на сумму даёт порядка $\delta$, т.к. из-за этого сумма и конечная.

И того относительная погрешность равна

$$\dfrac{e^x - \hat{e^x}}{e^x} \approx e^{|x|-x} |x| \epsilon_{маш}.$$

### Код

    #include<iostream>
    #include<cmath>
    #include<iomanip>

    using namespace std;

    int main(int argc, char *argv[] ){
        float i = 1, x ;
        cin >> x;
        bool not_possitive = false;
        if (x<0) {
            not_possitive = true;
            x = abs(x);
        }
        float sum = 1, term = x/i;

        while (sum != term + sum) {
            sum = term + sum;
            i++;
            term = term * x/i;
        }	

        if (not_possitive) {
            sum = 1/sum;
            x = -x;
        }
        cout << setprecision(26);
        cout << "Значение экспоненты:  "<< sum << endl;
        cout << "Отличие от экспоненты из библиотеки cmath:  " << abs(exp(x) - sum ) << endl;

        return 0;
    }
