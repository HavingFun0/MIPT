## Задача 6(9.7)


**Решение:**

а) Решение дифференциальной задачи:
$$u(x,t) = \int\limits_{-\infty}^{\infty} \frac{dy}{\sqrt{4 \pi t}} e^{-\frac{(x-y)^2}{4 t}} u(y,0) = \sum_{k=1}^\infty c_k \, e^{-{\pi}^2k^2t}\sin\left(k \pi x\right)$$

b) Переформулируем задачу в терминах нахождения собственной функции некоторого оператора:
$$u_m^{p+1} = (E + \tau \Lambda_{xx}) u_m^p$$
Будем искать решение в виде функции собственной функции оператора $ u_m^p = \sin\left( \frac{k \pi m}{M}\right)$

Подставляя в правую часть получаем с.з.

$$\lambda_k = 1 - \frac{4 \tau}{h^2} \sin^2\left(\frac{k \pi}{2 M}\right), k = 1,2,\dots, M-1.$$

Получим

$$u^p_m =  \sum_{k=1}^{M-1} \tilde{c}_k \, \left(1 - \frac{4 \tau}{h^2} \sin^2\left(\sqrt{2}\frac{\pi k}{2 M}\right) \right)^p \left( \sin\left(\frac{k \pi m}{M}  \right) \right)$$

Для нахождения коэффицентов $\tilde{c}_k$ учем начальное условие $y^0_m = \varphi(mh)$
Из скалярного произведения на сетке(просто усреднение по значениям в узлах) ясно, что
    $$\tilde{c}_k =  \dfrac{1}{M} \sum_{m = 0}^M \varphi(mh) \left(\sqrt{2} \sin\left(\dfrac{k \pi m}{M} \right) \right)$$


c) Теперь посмотрим как ведет себя схема при разных $k$:

1) Малые k

Раскладываем $\lambda_k$ по малости k получаем:

$$\left(1 - \tau \pi^2 k^2 \right)^{\frac{\pi^2 k^2 p \tau}{\pi^2 k^2 \tau}} \underset{\tau \to 0}\approx e^{-\pi^2 k^2 \tau} $$
Как видим всё ок

2) Большие k

Раскладываем $\lambda_k$ при k близких к M, получаем:
$$(-1 + 4r)^{\frac{t}{\tau}} \left(1 - \frac{\tau \pi^2}{1 - 4r} \right)^{\frac{\pi^2 p \tau}{\pi^2 \tau}\frac{1 - 4r}{1 - 4r}} \underset{\tau \to 0}\approx (-1 + 4r)^{\frac{t}{\tau}} e^{-\frac{\pi^2 t}{1- 4r}}$$

При $r < 0.5$ выражение стремиться к нулю, а при $r\geq 0.5$ -- бесконечно возрастает.

```python
x_i = -10
x_f = 10
t_i = 0
t_f = 5
dt = 0.002


# Количество узлов 
n = int(np.ceil((t_f - t_i)/dt))


# Временная сетка

time_grid = np.linspace(t_i, t_f, n)
```

```python
h_real = 0.001

x_grid_real = np.arange(x_i, x_f, h_real)

tau_real = 0.01

time_grid_real = np.arange(t_i, t_f, tau_real)

def phi (t, x) :
    return np.exp( - x ** 2 / (1 + 4 * t ) ) / np.sqrt (1 + 4 * t )


M = len(x_grid_real)

N = len(time_grid_real)
    
y_real = np.array([[0.0]*M]*N)
y_real[0] = np.exp(- x_grid_real**2)
for j in tqdm(range(N-1)):
    y_real[j+1] = phi(time_grid_real[j], x_grid_real)

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
x, t = np.meshgrid(x_grid_real, time_grid_real)
    
ax.plot_surface(np.array(x), np.array(t), np.array(y_real))
ax.set_xlabel('X')
ax.set_ylabel('T')
ax.set_zlabel('Y')
plt.show()   
```

Схема Кранка-Николсона:
$$\dfrac{\hat y - y}{\tau} = \dfrac{1}{2} \Lambda_{xx} \hat y + \dfrac{1}{2} \Lambda_{xx} y + \dfrac{1}{2}\left(f^{n+1}+f^n\right)$$

Сначала заменим вторую производную на разностную схему:

$$\frac{\partial y_m}{\partial t} = \frac{y_{m-1} - 2 y_m + y_{m+1}}{h^2}$$

Кравевые условия на границе:

$$u(x_{in},t) = \varphi_1(t), \ u(x_{f},t) = \varphi_2(t),$$
Теперь запишем матрицу A правой части уравнения:

$$\frac{\partial \vec y}{\partial t} = A \vec y$$

А затем применим неявный метод трапеции:
$$\frac{ \vec y^{n+1} -\vec y^{n}}{\tau} =\dfrac{1}{2} \left( A \vec y^{n+1} + A \vec y^{n} \right)$$

Выражая явно $\vec y^{n+1}$ получим
$$\vec y^{n+1} = \left(\mathbb{1} - \dfrac{\tau}{2} A\right)^{-1} \left(\mathbb{1} + \dfrac{\tau}{2} A\right) \vec y^n $$

Задачу будем решать на интервале $x\in[-5,5]$ и по времени $t\in[0,5]$. Сетку по времени возьмем $\tau = 0.0001$ 

```python
##Определим гран условия 
def phi_1 (t) :
    return np.exp( - x_i ** 2 / (1 + 4 * t ) ) / np.sqrt (1 + 4 * t )

def phi_2 (t) :
     return np.exp( - x_f ** 2 / (1 + 4 * t ) ) / np.sqrt (1 + 4 * t )
```

**Замечание:** Смотря на начальный профиль, понимая физику уравнения можно смело взять нулевые краевые условия и это не изменит сути решения, ибо точное решение при больших x $\approx 0$ при любых t.

С гран условиями первое и последнее уравнения будут иметь немного другой вид:
$$        \frac{\partial y_1}{\partial t} = \frac{\varphi_1(t) - 2 y_1 + y_2}{h^2}, \ \frac{\partial y_{N-1}}{\partial t} = \frac{y_{N-2} - 2 y_{N-1} + \varphi_2(t)}{h^2}.$$


Т.е. да у нас есть вектор правой части, который тоже нужно учесть:
$$\vec y^{n+1} = \left(\mathbb{1} - \dfrac{\tau}{2} A\right)^{-1} \left(\mathbb{1} + \dfrac{\tau}{2} A\right) \vec y^n + \dfrac{1}{2} \left(\mathbb{1} - \dfrac{\tau}{2} A\right)^{-1} \left(\vec f^{n+1}+\vec f^n\right)$$

где $\vec f^n = (\varphi_1^n, 0,...,0,\varphi_2^n)^T$.

Будем решать для разных шагов h. Обращать матрицу будем методом Зейделя с точностью $\epsilon\leq 0.02$

```python
def seidel(A, b, eps):
    n = len(A)
    x = np.zeros(n)  

    converge = False
    while not converge:
        x_new = np.copy(x)
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]

        converge = np.sqrt(sum((x_new[i] - x[i]) ** 2 for i in range(n))) <= eps
        x = x_new

    return x
```
