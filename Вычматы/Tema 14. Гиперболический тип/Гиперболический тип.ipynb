Предиктор:
$$   \frac{\tilde{y}_m - y_m^n}{\tau} + \frac{f_{m+1}^n - f_m^n}{h} = 0,$$
$$    \frac{\tilde{y}_{m-1} - y_{m-1}^n}{\tau} + \frac{f_m^n - f_{m-1}^n}{h} = 0$$

$$    \frac{y_m^{n+1} - \frac{y_m^n + \tilde{y}_m}{2}}{\tau} + \frac{\tilde{f}_m - \tilde{f}_{m-1}}{2 h} = 0$$

$$\lambda = 1 + \sigma^2 (\cos(\alpha) - 1) - i \sigma \sin \alpha$$

```python
# Начаные условия
def init_cond(x, a):
    return (a*x**2+1.1)*np.sin(np.pi*x)

for h in hs:
    for dt in dts:
        N = int((x_f-x_i)/h)
        M = int((t_f-t_i)/dt)
        space_grid = np.linspace(x_i, x_f, M)
        time_grid = np.linspace(t_i, t_f, N)

        u = np.zeros([N,M])
        u[0] = init_cond(space_grid, 1.1)
        #u[]
        u[1] = u[0]
        #u[]
        
        for i in range(1,N-1):
            for j in range(1,M-1):
                u[i+1,j] = 2*u[i,j] - u[i-1,j] + dt**2 / h**2 *(u[i, j+1]-2*u[i,j] + u[i,j-1])
        fig = plt.figure()
        ax = fig.add_subplot(111, projection = '3d')
        x, t = np.meshgrid(space_grid, time_grid)
    
        ax.plot_surface(np.array(x), np.array(t), np.array(u))
        ax.set_xlabel('X')
        ax.set_ylabel('T')
        ax.set_zlabel('U')
        ax.set_title('h = '+str(h)+' , dt = '+ str(dt))
        ax.view_init(40, 30)
        plt.show()   
        #print(u[])
        #print(u[N-5])
```

Точное решение будем искать в виде ряда(просто по другому не знаю как)
$$u(x,t) =\sum_{n=1}^{\infty} \beta_n \cos(\pi n t) \sin(\pi n x)$$
где $\beta_n = 1.1 \delta(n-1) - \dfrac{8 a n (-1)^n}{(n^2-1)^2}$ -- коэффицент разложения $u(x,0)$ по синусам $\sin(\pi n x)$ на [0,1].

```python
def u(x,t):
    s=0
    for i in tqdm(range(1,10**6)):
        s+= -8*1.1*(i+1)*(-1)**(i+1)/((i+1)**2-1)**2 * np.cos(np.pi * (i+1)*t)*np.sin(np.pi*x*(i+1))
    return np.sin(np.pi*x)+ s
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
    
ax.plot_surface(np.array(x), np.array(t), u(x, t))
ax.set_xlabel('X')
ax.set_ylabel('T')
ax.set_zlabel('U')
ax.set_title('h = '+str(h)+' , dt = '+ str(dt))
ax.view_init(40, 30)
plt.show() 
```
