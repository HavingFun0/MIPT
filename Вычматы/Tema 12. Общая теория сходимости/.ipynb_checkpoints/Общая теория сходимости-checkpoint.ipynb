## Задача 1(7.1)


**Решение:**

а) $U =0$

    1) Аппроксимация 
    Разложим в ряд Тейлора обе части получим


Исходя из условия $\psi$ -- решение уравнения $- i \dfrac{\partial \psi}{\partial t} = \dfrac{\partial^2 \psi}{\partial x^2}$. Сокращая это уравнение получем аппроксимацию $\omicron(dt, dx^2)$

    2) Устойчивость.
    Проверим устойчивость по признаку Неймана
$\psi^n_m = \lambda^n e^{i \alpha m}$


Из второго выражения видно, что для того, чтобы $|\lambda|<=1$ нужно чтобы модуль мнимой части числителя был меньше модуля мнимой части знаменателя

Итог $\xi\geq 1/2$.

б) $U\ne 0$

    1) Аппроксимация
    Проверяется тем же способом и получается тот же результат
    2) Устойчивость.
    Аналогично получаем выражение
    

Как видно ничего не поменялось(имеется ввиду, что случай свелся к предыдущему). Следовательно $\xi\geq 1/2$.

Теорема о сходимости говорит, что сходимость = аппроксимация + устойчивость.

## Задача 2(7.2)


**Решение:**

Ищем в виде $a \psi^{n+1}_m + b\psi^{n}_{m-1}+k\psi^{n}_{m}+d\psi^{n}_{m+1}=0$
Раскладиваем в ряд Тейлора и используем, что $\psi$ удовлетворяет уравнения переноса получаем(т.е. подбираем коэффиценты так, чтобы оно осталось). Получаем систему уравнений:

\begin{cases}
      a+b+k+d = 0      \\
      (d-b)h = c\\
      a dt = 1\\
      a c^2 dt^2+(b+d) h^2 =0
\end{cases}

Отсюда $a = \dfrac{1}{dt}, ~ b = \dfrac{-c^2 dt - c h}{2 h^2}, ~ d= \dfrac{-c^2 dt + c h}{2 h^2}, ~k = \dfrac{c^2 dt}{h^2}-\dfrac{1}{dt}.$


Итог: $$\frac{y_{m}^{n+1} - y_{m}^n}{\tau} + \frac{c}{2h} \left( y_{m+1}^n - y_{m-1}^n \right) -
\frac{c^2 \tau}{2h^2}\left( y_{m+1}^n-2 y_m^n+y_{m-1}^n \right) = 0
.$$

Порядок аппроксимации схемы -- $\omicron(dt^2,dx^2).$

Исследуем на устойчивость методом Неймана.

Вводя замену $\sigma = \dfrac{c dt}{dx}$ получим
$$\lambda = 1 - i\sigma \sin \alpha-\sigma^2 (1-\cos \alpha)$$

Теперь нужно посмотреть, чтобы $|\lambda|^2<=1$.

Что выполняется при $\sigma^2 <=1.$ Следовательно схема устойчива.

## Задача 3(7.3)


**Решение:**
Абсолютно аналогично:

Ищем в виде $a \psi^{n+1}_{m} + b\psi^{n}_{m}+k\psi^{n}_{m-1}+d\psi^{n}_{m-2}=0$
Раскладиваем в ряд Тейлора и используем, что $\psi$ удовлетворяет уравнения переноса получаем(т.е. подбираем коэффиценты так, чтобы оно осталось и его дифференциальные следствия). Получаем систему уравнений:


\begin{cases}
      a+b+k+d = 0      \\
      (-2d -k)h = c\\
      a dt = 1\\
      \dfrac{a}{2} c^2 dt^2+\dfrac{4d + k}{2} h^2 =0
\end{cases}

Решая её получим     
\begin{cases}
        \gamma = -\frac{2c}{h}+\frac{c^2\tau}{h^2},
        \\
        \alpha = \frac{1}{\tau},
        \\
        \delta = \frac{c}{2h}- \frac{c^2 \tau}{2 h^2},
        \\
        \beta = -\frac{1}{\tau}+\frac{3c}{2h}- \frac{c^2 \tau}{2h^2}.
    \end{cases}
    
    
Итог:
$$\frac{y_m^{n+1} - y_m^n}{\tau} + \frac{c}{2h} \left( 3 y_m^n - 4 y_{m-1}^n + y_{m-2}^n\right) - \frac{c^2 \tau}{2h^2} \left( y_m^n - 2y_{m-1}^n + y_{m-2}^n\right) = 0$$


Исследуем на устойчивость методом Неймана.

Вводя замену $\sigma = \dfrac{c dt}{dx}$ получим

$\lambda = \sigma(\sigma -2) + (\sigma-1)^2 \cos(\alpha) + i (\sigma-1) \sin(\alpha)$
Условие $|\lambda|^2\leq 1$ выполняется при $\sigma \leq 2$.

## Задача 5(7.8)


**Решение:**

Уравнение Лапласа:

$$\Delta u =0$$

Схема будет выглядеть так:


Разностная схема и её шаблон


$$a y_{m+1,\,k+1}+b y_{m+1,\,k-1}+c y_{m-1,\,k+1}+d y_{m-1,\,k-1}+e y_{m,\,k}=0$$

 Получаем систему уравнений:

\begin{cases}
a + b + c + d + e = 0 \\
a + b - c - d = 0 \\
a - b + c - d = 0 \\
a - b - c + d = 0 \\
\frac{h^2}{2} ( a+b+c+d) =1
\end{cases}


Решая систему получим:

$$a = b = c = d = \frac{1}{2h^2}; \quad e = -\frac{2}{h^2}$$

Итог:
\begin{equation*}
\frac{y_{m+1,\,k+1}-2 y_{m,\,k}+y_{m-1,\,k-1}}{2h^2}+ \frac{y_{m-1,\,k+1}-2y_{m,\,k}+y_{m+1,\,k-1}}{2h^2}=0
.\end{equation*}

## Задача 8(7.28з)



**Решение:**
Всё тоже самое:

1) Аппроксимация.

Раскладывая в ряд Тейлора и применяя дифференциальные следствия получаем порядок аппроксимации $\omicron(dt^2,dx^2)$

Устойчивость проверяем всё так же по признаку фон Неймана:
$$ \frac{\lambda - 1}{\tau} = \lambda\frac{ e^{i\alpha} - 2 + e^{-i\alpha} }{2h^2} + \frac{ e^{i\alpha} - 2 + e^{-i\alpha} }{2h^2} = (\lambda + 1)\frac{ \cos(\alpha) - 1 }{h^2}$$

Получаем:
$$\lambda = \frac{ 1 - 2\sigma\sin^2(\alpha/2) }{ 1 + 2\sigma\sin^2(\alpha/2) },~~ |\lambda| < 1 ~~\forall \alpha$$

Следовательно схема сходится к решению диффура.
