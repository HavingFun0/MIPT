## Задача 1

Дифференциальная задача Коши легко решается:
\begin{equation*}
    u'_{t} = -3u, u(0) = 3 \Rightarrow u(t) = 3 \exp(-3t)    
\end{equation*}
Решим разностную задачу:
\begin{equation*}
\begin{gathered}
    \frac{y_{n+1}-y_{n}}{\Delta t} = 0.6f(t_{n+1},y_{n+1})+0.4f(t_{n},y_{n})\\
    \frac{y_{n+1}-y_{n}}{\Delta t} = 0.6f(-3y_{n+1})+0.4(-3y_{n})\\
    y_{n+1}-y_{n} = \Delta t(-1.8y_{n+1}-1.2y_{n})\\
    y_{n+1}(1+1.8\Delta t) = y_{n}(1-1.2\Delta t)\\
    y_{n+1} = y_{n} \frac{1-1.2\Delta t}{1+1.8\Delta t}
\end{gathered}
\end{equation*}
Из полученного рекурентного соотношения получаем формулу n-го члена
\begin{equation*}
    y_{n} = y_{0} (\frac{1-1.2\Delta t}{1+1.8\Delta t})^{n} = 3 (\frac{1-1.2\Delta t}{1+1.8\Delta t})^{n} = 3 (1 + \frac{-3\Delta t}{1+1.8\Delta t})^{n}
\end{equation*}
Пусть мы ищем решение на отрезке [0, t], тогда $\Delta t = \frac{t}{n}$
\begin{equation*}
    \lim\limits_{n\to \inf} 3 (1 + \frac{-3\frac{t}{n}}{1+1.8\frac{t}{n}})^{n} = 3\lim\limits_{n\to \inf} (1 + \frac{-3t}{n+1.8t})^{n}
\end{equation*}
Сделаем замену $m = n +1.8t$, тогда
\begin{equation*}
\begin{gathered}
    3\lim\limits_{n\to \inf} (1 + \frac{-3t}{n+1.8t})^{n} = 3\lim\limits_{m\to \inf} (1 + \frac{-3t}{m})^{m+1.8t} =\\
    = 3\lim\limits_{m\to \inf} (1 + \frac{-3t}{m})^{m} \cdot \lim\limits_{m\to \inf} (1 + \frac{-3t}{m})^{1.8t} = 3\lim\limits_{m\to \inf} (1 + \frac{-3t}{m})^{m} \cdot 1 = 3 exp(-3t)
\end{gathered}
\end{equation*}
Следовательно
\begin{equation*}
    \lim\limits_{n\to \inf} ||u_{n} - y_{n}|| = 0
\end{equation*}
То есть решение разностной задачи сходится к решению дифференциальной.
\newpage
Найдем порядок сходимости:
\begin{equation*}
\begin{gathered}
    r := \frac{u_{n+1} - u_{n}}{\Delta t} - (0.6f(u_{n+1})+0.4f(u_{n})) =\\
    = \frac{u_{n+1} - u_{n}}{\Delta t} - (0.6u'_{n+1}+0.4u'_{n}) =\\
    = \frac{u_{n}+u'_{n}\Delta t+\frac{1}{2}u''(\Theta_{1})\Delta t^{2} - u_{n}}{\Delta t} - (0.6(u'_{n} + u''(\Theta_{2}) \Delta t)+0.4u'_{n}) =\\
    = (u''(\Theta_{1}) + 0,6u''(\Theta_{2})) \Delta t
\end{gathered}
\end{equation*}
Следовательно
\begin{equation*}
r \leq 1,6 \cdot M_{4} \cdot \Delta t
\end{equation*}
Первый порядок сходимости.

\begin{equation*}
\begin{gathered}
    -u_{n+1}+3u_{n}-2u_{n-1}=dt(au'_{n+1}+bu'_{n}+cu'_{n-2})\\
    r := \frac{-u_{n+1}+3u_{n}-2u_{n-2}}{dt}-(au'_{n+1}+bu'_{n}+cu'_{n-2})\\
\end{gathered}
\end{equation*}
Используя
\begin{equation*}
\begin{gathered}
    u_{n+1} = u_{n}+u'_{n}dt+\frac{1}{2}u''_{n}dt^{2}+\frac{1}{6}u'''_{n}dt^{3}+\frac{1}{24}u''''_{n}dt^{4}\\
    u_{n-2} = u_{n}+u'_{n}(-dt)+\frac{1}{2}u''_{n}(-dt)^{2}+\frac{1}{6}u'''_{n}(-dt)^{3}+\frac{1}{24}u''''_{n}(-dt)^{4}\\
    u'_{n+1} = u'_{n}+u''_{n}dt+\frac{1}{2}u'''_{n}dt^{2}+\frac{1}{6}u''''_{n}dt^{3}\\
    u'_{n-2} = u'_{n}+u''_{n}(-2dt)+\frac{1}{2}u'''_{n}(-2dt)^{2}+\frac{1}{6}u''''_{n}(-2dt)^{3}\\
\end{gathered}
\end{equation*}
Получаем
\begin{equation*}
\begin{gathered}
    r = u'_{n}-\frac{3}{2}u''_{n}dt+\frac{1}{6}u'''_{n}dt^{2}-\frac{3}{24}u''''_{n}dt^{3}-((a+b+c)u'_{n}+(a-2c)u''_{n}dt+\frac{1}{2}(a+4c)u'''_{n}dt^{2}+\frac{1}{6}(a-8c)u''''_{n}dt^{3})
\end{gathered}
\end{equation*}
Чтобы получить схему наибольшего порядка аппроксимации необходимо потребовать
\begin{equation*}
\begin{cases}
    a+b+c = 1\\
    a-2c = -\frac{3}{2}\\
    a+4c = \frac{1}{3}
\end{cases}
\end{equation*}
Решением системы являются значения параметров
\begin{equation*}
\begin{cases}
    a = -\frac{8}{9}\\
    b = \frac{19}{12}\\
    c = \frac{11}{36}
\end{cases}
\end{equation*}
Таким образом
\begin{equation*}
\begin{gathered}
    r = -\frac{3}{24}u''''_{n}dt^{3} - \frac{1}{6}(a-8c)u''''_{n}dt^{3} =\frac{93}{216}u''''_{n}dt^{3}\\
    ||r|| \leq \frac{93}{216} \cdot M_{4} \cdot dt^{3}
\end{gathered}
\end{equation*}
Максимальный порядок аппроксимации третий.\\

Теорема (второй барьер Далквиста): Любой А-устойчивый многошаговый метод должен иметь порядок аппроксимации не выше второго. Если порядок равен 2, то константа погрешности метода удовлетворяет неравенству $C \leq - \frac{1}{12}$.

Из данной теоремы очевидно следует что найденный многошаговый метод не является А-устойчивым.
