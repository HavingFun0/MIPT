## Задача 1(11.5)

a) Рассмотрим функцию $g(x) = \dfrac{f(x)}{f'(x)}$. Она имеет корень корень кратности 1 в той же точке, что и $f(x)$.

Рассмотрим метод Ньютона как итерационный процесс: $x^{n+1} = F(x^n)$, где $F(x) = x - \dfrac{g(x)}{g'(x)}$. Решение уравнения -- неподвижная точка $F(x^*) = x^*$. Поэтому 

$$x^{n+1}-x^* = F(x^n) - F(x^*) \approx F'(x^*)(x^n - x^*)+ \dfrac{F''(x^*)}{2} (x^n - x^*)^2 + ...$$

Считая производные и подставляя в них точку корня найдем, что $F'(x^*) = 0$, $F''(x^*) = \dfrac{g''(x^*)}{g'(x^*)}\neq 0$. Следовательно Ньютона для $f(x)$:

$$x_{n+1} = x_n - \dfrac{f(x_n)f'(x_n)}{(f'(x_n))^2-f(x_n) f''(x_n)}$$

b) Рассмотрим метод 
$$x_{n+1} = x_n - m\dfrac{f(x_n)}{f'(x_n)}$$
где $m={1,2,..}$. Будем для одинаковых начальных точек запускать метод и смотреть за сколько итераций он сойдется. Максимальная(квадратичная) скорость сходимости будет, когда m=p. Добавление множителя, равного кратности корня, означает, что фактически вместо уравнения $f(x)=0$ мы пытаемся решать уравнение $\sqrt[p]{f(x)}=0$ (точностью до знаков), для которого первая производная в окрестности корня уже ненулевая. Поэтому добавление этого множителя восстанавливает квадратичную скорость сходимости.

**Решение**

Очевидно, что будет сходится, если $\phi(x)$ будет сжимающим отображением.

Для этого достаточно, чтобы $|\phi'(x)| < 1$:

$$|\phi'(x)| = |2 a b^2 x \exp(-b^2 x^2)| <1$$

$$|\phi'(x)| = |2 a b^2 x \exp(-b^2 x^2)| < \dfrac{2}{\sqrt{2 e}}| a b| < 1 \Rightarrow$$
$$ |ab| < \sqrt{e/2}$$

## Задача 4(12.4е)

Корень лежит в интервале от (2.7,2.8). Найдем его используя метод простой итерации.

На этом интервале отображение $\arcsin \left( \dfrac{x^2}{20} \right)$ является сжимающем и имеет максимум подуля производной равный $\approx$ 0.3359355.

Для уточнения корня устроим итерационный процесс до точности $10^{-5}$, для этого понадобится $n = \dfrac{\log(10^{-5})}{\log(0.3359355) \approx 11 }$ итераций.

```python
import numpy as np
```

```python
x = 2.8
for i in range(11):
    x = np.arcsin(x*x/20)

x
```

Как видим, метод простой итерации находит только тривиальны корень, что достаточно понятно, так как метод простой итерации ищет стационарную точку отображения, а она существует и единственна и, очевидно, равна 0. 

Попробуем метод Ньютона.

```python
from scipy import misc


def f(x):
    return -np.sin(x)*20+x**2

x = 2.8


while True:
    x1 = x - f(x) / misc.derivative(f, x) 
    t = abs(x1 - x)
    if t < 10**(-5):
        break
    x = x1
    
x
```

Всё отлично сработало.

## Задача 5(12.7д)

Как видим их всего 5 штук. Найдем их с помощью итерационного процесса.

Для начала выразим так:
$$x = \sqrt[4]{\dfrac{y^3-105}{3 y}} = f(x,y), ~~~ y =-\sqrt[4]{\dfrac{x^7 +1510}{5 x^2}} = g(x,y)$$

```python
def f(x,y):
    return ((y**3-105)/3.0/y)**(1/4)

def g(x,y):
    return -((x**7+1510)/5.0/x**2)**(1/4)
```

**Окрестность x = -2**

```python
x_in = -2
y_in = -3
x =0
y = 0

while (np.linalg.norm([x-x_in,y-y_in]) > 10**(-5)):
    x = x_in
    y = y_in
    x_in = -f(x,y)
    y_in = g(x,y)
    print('X_in = ' + str(x_in)+ '  Y_in = ' + str(y_in))
    
print('\n\nx = ' + str(x_in) +', y = ' + str(y_in))
```

**Окрестность x = 2**

```python
x_in = 2
y_in = -3
x =0
y = 0

while (np.linalg.norm([x-x_in,y-y_in]) > 10**(-5)):
    x = x_in
    y = y_in
    x_in = f(x,y)
    y_in = g(x,y)
    print('X_in = ' + str(x_in)+ '  Y_in = ' + str(y_in))
    
print('\n\nx = ' + str(x_in) +', y = ' + str(y_in))
```

**Окрестность x = -1**

Выразим по-другому:
$$x = \sqrt{\dfrac{x^7+1510}{5y^4}} = f(x,y), ~~~ y =\sqrt[3]{3 x^4 y+105} = g(x,y)$$

```python
def f(x,y):
    return np.sqrt((x**7 + 1510)/5/y**4)


def g(x,y):
    q = 3 * x**4 * y + 105
    if (q>=0):
        return q**(1/3)
    else:
        return -(-q)**(1/3)
```

**Окрестность х = -1**

```python
x_in = -1
y_in = 5
x =0
y = 0

while (np.linalg.norm([x-x_in,y-y_in]) > 10**(-5)):
    x = x_in
    y = y_in
    x_in = -f(x,y)
    y_in = g(x,y)
    print('X_in = ' + str(x_in)+ '  Y_in = ' + str(y_in))
    
print('\n\nx = ' + str(x_in) +', y = ' + str(y_in))
```

**Окрестность х = 1**

```python
x_in = 1
y_in = 5
x =0
y = 0

while (np.linalg.norm([x-x_in,y-y_in]) > 10**(-5)):
    x = x_in
    y = y_in
    x_in = f(x,y)
    y_in = g(x,y)
    print('X_in = ' + str(x_in)+ '  Y_in = ' + str(y_in))
    
print('\n\nx = ' + str(x_in) +', y = ' + str(y_in))
```

**Окрестность x = -3**

Выразим по-другому:
$$x = \sqrt[7]{5 x^2 y^4-1510} = f(x,y), ~~~ y =\dfrac{y^3-105}{3 x^4} = g(x,y)$$

```python
def f(x,y):
    q =(5* x**2 *y**4 - 1510)
    if (q>=0):
        return (q)**(1/7)
    else:
        q = -q
        return -1*(q)**(1/7)


def g(x,y):
    return (y**3-105)/3.0/x**4


x_in = -3
y_in = -1
x =0
y = 0

while (np.linalg.norm([x-x_in,y-y_in]) > 10**(-5)):
    x = x_in
    y = y_in
    x_in = f(x,y)
    y_in = g(x,y)
    print('X_in = ' + str(x_in)+ '  Y_in = ' + str(y_in))
    
print('\n\nx = ' + str(x_in) +', y = ' + str(y_in))
```
