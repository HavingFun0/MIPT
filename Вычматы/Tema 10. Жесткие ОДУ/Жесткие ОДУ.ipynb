## Задача 9(8.3)


```python
#Найдём собственные вектора матрицы ОДУ
a = np.array([[998,1998],[-999,-1999]])

LA.eig(a)[0]
```

Собственные числа матрицы: $$\lambda = -1, \lambda = -1000.$$ 

Точное решение: 

$$u = 4e^{-x} - 3e^{-1000x},~ v = -2e^{-x} + 3e^{-1000x}.$$

**1) Явный метод Эйлера**


Функция устойчивости: $R(z) = 1 + \lambda h \Rightarrow h \leqslant 2\cdot10^{-3}$

```python
N = 10001
h = 0.0019
u, v = np.zeros(N), np.zeros(N)
u[0], v[0] = 1,1 #initial conditions

for i in range(1, N):
    u[i], v[i] = np.array([u[i-1], v[i-1]]) + h*a@np.array([u[i-1], v[i-1]])
```

**2) Неявный метод Эйлера**

Т.к. $R(z) = \dfrac{1}{1-z},$ то $h \geq 0.002$


```python
from tqdm import tqdm
```

```python
h = 10**(-5)
N1 = int(10//h + 1)
u1, v1 = np.zeros(N1), np.zeros(N1)
u1[0], v1[0] = 1,1 #initial conditions

for i in tqdm(range(1, N1)):
    u1[i], v1[i] = np.linalg.inv(np.eye(2) - h*a)@(np.array([u1[i-1],v1[i-1]]))
```

```python
x = np.linspace(0, 10, N)
x1 = np.linspace(0, 10, N1)

plt.plot(x, 4*np.exp(-x) - 3*np.exp(-1000*x), label = 'Точное')
plt.plot(x, u, label = 'Явный')
plt.plot(x1, u1, label = 'Неявный')

plt.title('График решений')
plt.ylabel('y')
plt.xlabel('x')
plt.grid()
plt.legend()
plt.show()
```
