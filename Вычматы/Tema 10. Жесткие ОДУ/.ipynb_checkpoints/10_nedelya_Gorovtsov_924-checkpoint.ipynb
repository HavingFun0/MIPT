Ну раз можно пользоваться теоремой для случая, когда число стадий $s$ совпадает с порядком аппроксимации $p$, то функция устойчивости это разложение экспоненты до порядка $s = \overline{1,7}$:
$$ R(z) = 1 + z +... + \frac{z^p}{p!} $$

Двухстадийный метод: 
$$ y_{n+1} = y_n + h(b_1k_1 + b_2k_2) $$
$$ k_1 = f(x_n + c_1h, y_n + h(a_{11}k_1 + a_{12}k_2)) $$
$$ k_2 = f(x_n + c_2h, y_n + h(a_{21}k_1 + a_{22}k_2)) $$

Первый порядок: $y_{n+1} = y_n + y'(x_n)h = y_n + f(x_n,y_n)h$. Так как мы работаем в первом порядке, то функцию $f$ считаем не изменившейся и получаем, что вполне достаточно требовать $b_1 + b_2 = 1$

Второк порядок: $y_{n+1} = y_n + y'(x_n)h + \frac{y''(x_n)h^2}{2}$. Тут уже придётся раскладывать $f$:
$$ f(x_n + c_1h, y_n + h(a_{11}k_1 + a_{12}k_2)) = y'(x_n) + \frac{\partial f}{\partial x}c_1h + \frac{\partial f}{\partial y}(a_{11}k_1 + a_{12}k_2)h $$
$$ f(x_n + c_2h, y_n + h(a_{21}k_1 + a_{22}k_2)) = y'(x_n) + \frac{\partial f}{\partial x}c_2h + \frac{\partial f}{\partial y}(a_{21}k_1 + a_{22}k_2)h $$

Члены с производной по $x$ нам нужны, с производной по $y$ - нет:
$$ b_1c_1 + b_1c_2 = 1/2 $$
$$ b_1a_{11} + b_2a_{21} = 0 = b_1a_{12} + b_2a_{22} $$

Ищем решение уравнения Далквиста ($f_n = \lambda y_n$) в виде $y_n \propto q^n$:

$$ z(q) = \frac{3 - 2/q + 1/q^2}{2} $$

Надо посмотреть, куда переходит граница единичного круга $q = e^{i\varphi}$: 

$$ z(\varphi) = \frac{3 - 2e^{-i\varphi} + e^{-2i\varphi}}{2} = 1 - 2\sin^2(\frac{\varphi}{2})e^{i\varphi} $$

```python
angle = np.linspace(0, 2*np.pi, 100)
x = 1 - 2*((np.sin(angle/2))**2)*np.cos(angle)
y = -2*((np.sin(angle/2))**2)*np.sin(angle)
plt.plot(x, y)
```

Учитывая то, что $q = 0$ переходит в бесконечность, нам нужна внешность этой ... фигуры

Условия Кутты: $a_{11} + a_{12} = c_1 \Rightarrow a_{11} = \frac{1}{2} - \frac{\sqrt{3}}{6}$
В однократно диагонально-неявном методе $a_{11} = a_{22} \Rightarrow c_2 = \frac{1}{2} + \frac{\sqrt{3}}{6}$.

На коэффициенты $b_i$ у нас есть два условия: 
\begin{equation}
\begin{cases}
    b_1 + b_2 = 1 \\
    b_1c_1 + b_2c_2 = \frac{1}{2}
\end{cases}
\end{equation}

Получаем $b_1 = b_2 = \frac{1}{2}$. 

Условия третьего порядка аппроксимации:

$b_1c_1^2 + b_2c_2^2 = \frac{2}{8} + \frac{2}{24} = \frac{1}{3}$

$ b_1(a_{11}c_1 + a_{12}c_2) + b_2(a_{21}c_1 + a_{22}c_2) = \frac{1}{6} $

Для четвертого порядка:

$ \sum_{i=1}^2 a_{ij} b_i c_i = \dfrac{b_j}{2} (1-c_j^2)$


Для $j=1$: $\left( \frac{1}{2} - \frac{\sqrt{3}}{6} \right)^2\frac{1}{2} \ne \frac{1}{4}\left( 1 - \left( \frac{1}{2} - \frac{\sqrt{3}}{6} \right) \right)^2 = \frac{1}{4}\left( \frac{2}{3} + \frac{\sqrt{3}}{6} \right)$

Функция устойчивости: 

$R(z) = \frac{ \left(1+\frac{\sqrt{3z}}{6}\right)^2 -\frac{z}{2}\frac{z(\sqrt{3}-2)}{2\sqrt{3}} }{\left(1 - \left( \frac{1}{2} - \frac{\sqrt{3}}{6} \right)z \right)^2} = \frac{1 + \frac{ \sqrt{3} }{3}z - \frac{z^2}{6} + \frac{z^2}{2\sqrt{3}} }{1 - \left( 1 - \frac{1}{\sqrt{3}} \right)z + \left( \frac{1}{3} - \frac{\sqrt{3}}{6} \right)z^2}$

Метод точно не $L$-устойчив, так как степени числителя и знаменателя равны. $R(z) \to \frac{\sqrt{3} - 1}{2 - \sqrt{3}} > 1, \, Re(z) \to -\infty$, следовательно метод не $A$-устойчив. 

```python
x = np.linspace(-100, 0, 100)
y = ((1 + x*np.sqrt(3)/6)**2 + (x**2)*(2 - np.sqrt(3))/4/np.sqrt(3))/((1 - (1/2 - np.sqrt(3)/6)*x)**2)

plt.plot(x, y)
plt.hlines( (np.sqrt(3) - 1)/(2-np.sqrt(3)) , -100, 0, color = 'red')
plt.title('Метод монотонный')
```

Для неявной $R(z) = \frac{1}{1-z}$ - метод монотонный

Схема Розенброка: $R(z) = \frac{2}{1 + (z-1)^2}$ - метод монотонный

Собственные числа матрицы: $\lambda = -1, \lambda = -1000$. Точное решение: $u = 4e^{-x} - 3e^{-1000x}$, $v = -2e^{-x} + 3e^{-1000x}$.

## Явно
Функция устойчивости: $R(z) = 1 + \lambda h \Rightarrow h \leqslant 2\cdot10^{-3}$

```python
A = np.array([[998, 1998], [-999, -1999]])
N = 10001
u, v = np.zeros(N), np.zeros(N)
u[0], v[0] = 1,1 #initial conditions

for i in range(1, N):
    u[i], v[i] = np.array([u[i-1], v[i-1]]) + 0.0019*A@np.array([u[i-1], v[i-1]])
```

## Неявно
Функция устойчивости уже требует $h \geqslant 0.002$

```python
h = 0.1
N1 = int(10//h + 1)
u1, v1 = np.zeros(N1), np.zeros(N1)
u1[0], v1[0] = 1,1 #initial conditions

for i in range(1, N1):
    u1[i], v1[i] = np.linalg.inv(np.eye(2) - h*A)@(np.array([u1[i-1],v1[i-1]]))
```

```python
x = np.linspace(0, 10, N)
x1 = np.linspace(0, 10, N1)

plt.plot(x, 4*np.exp(-x) - 3*np.exp(-1000*x), label = 'Точное', linewidth = 10)
plt.plot(x, u, label = 'Явный')
plt.plot(x1, u1, label = 'Неявный')

plt.legend()
```
