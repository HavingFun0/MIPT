# Задача 6.4


Правило Рунге:
$$I \approx I^p\left(\dfrac{h}{2}\right)+\dfrac{1}{2^p-1}\left(I^p\left(\dfrac{h}{2}\right)-I^p(h)\right)$$
где p -- порядок аппроксимации. Для метода трапеции p=2.
Также выпишем выражение для метода трапеции:
$$I^T\left(\dfrac{h}{2}\right) = \dfrac{h}{4}\sum_{k=0}^{n-1}(f(x_{k+1})+2f(x_k)+ f(x_{k-1}))$$


$$I^T\left(h\right) = \dfrac{h}{2}\sum_{k=0}^{n-1}(f(x_{k+1})+ f(x_{k-1}))$$
Подставляя всё в формулу Рунге:
$$I^p\left(\dfrac{h}{2}\right)+\dfrac{1}{2^p-1}\left(I^p\left(\dfrac{h}{2}\right)-I^p(h)\right) =
\dfrac{h}{4}\sum_{k=0}^{n-1}(f(x_{k+1})+2f(x_k)+ f(x_{k-1})) + \dfrac{1}{3 \cdot 4} *\sum_{k=0}^{n-1}(f(x_{k+1})+2f(x_k)+ f(x_{k-1}) - 2f(x_{k+1}) - 2 f(x_{k-1}) ) = \dfrac{h}{2} \dfrac{1}{6} \sum_{k=0}^{n-1}(2 f(x_{k+1}) + 2 f(x_{k-1}) + 8f(x_k) ) = \dfrac{h}{2} \dfrac{1}{3} \sum_{k=0}^{n-1}( f(x_{k+1}) +  f(x_{k-1}) + 4f(x_k) ) = I^{Simp}$$

Что и требовалось.

$$ f(x) \approx \frac{-x}{\sqrt{5x}x\sqrt{1 - \frac{x}{5}}} = \frac{-1}{\sqrt{5x}}\left( 1 + \frac{3 x}{5} \right)$$
Второе слагаемое нормальное, а вот первое даст нам особенность и нужно численно итегрировать $f(x) + \frac{1}{\sqrt{5x}}$ с точностью $\epsilon/2$.

Надо интегрировать $f(x) - \frac{1-e^5}{5\sqrt{5}(x-5)}$ c точностью $\epsilon/2$. 

После численного интегрирования нужно добавить аналитический вычисленные интегралы от $\frac{1}{\sqrt{5x}}$ и $\frac{1-e^5}{5\sqrt{5}(x-5)}$.

# Задача 9.5в


```python
x = np.array([0 + 0.15*i for i in range(9)])
y = np.array([1, 1.007568, 1.031121, 1.073456, 
              1.140228, 1.242129, 1.400176, 1.660300, 2.143460])

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
def its_a_trap(x,y):
    res = 0
    for i in range(len(x) - 1):
        res += (x[i+1]-x[i])*(y[i] + y[i+1])/2
    return res

def Simpson(x,y):
    res = 0
    for i in range(0, len(x) - 2, 2):
        res += (x[i+2]-x[i])*(y[i] + 4*y[i+1] + y[i+2])/6
    return res
```

```python
print('Трапеции по всем точкам: ', its_a_trap(x,y), '\nСимпсон: ', Simpson(x,y))
```

Теперь уточним интеграл с помощь. экстраполяции Ричардсона:
Посчитаем интеграл по всем точкам (это $I(h/2)$), посчитаем через одну (это $I(h)$) и оценим интеграл, как (формула трапеций имеет второй порядок аппроксимации)
$$ I \approx I(h/2) + \frac{I(h/2) - I(h)}{3} $$

```python
I1, I2 = its_a_trap(x[::2], y[::2]), its_a_trap(x,y)
print('I(h) = ', I1, '\nI(h/2) = ', I2, '\nI = ', I2 + (I2-I1)/3)
```

Как видно, полученные результаты совпадают

# Задача 9.13б



Пусть через первые три точки проходит один кубический многочлен, через последние три другой, а посередине мы сошьём их вплоть до второй производной:

\begin{equation}
\begin{cases}
    a_1x_0^3 + b_1x_0^2 + c_1x_0 + d_1 = f_0\\
    a_1x_1^3 + b_1x_1^2 + c_1x_1 + d_1 = f_1\\
    a_1x_2^3 + b_1x_2^2 + c_1x_2 + d_1 = f_2\\
    a_2x_2^3 + b_2x_2^2 + c_2x_2 + d_2 = f_2\\
    a_2x_3^3 + b_2x_3^2 + c_2x_3 + d_2 = f_3\\
    a_2x_4^3 + b_2x_4^2 + c_2x_4 + d_2 = f_4\\
    3a_1x_2^2 + 2b_1x_2 + c_1 = 3a_2x_2^2 + 2b_2x_2 + c_2\\
    6a_1x_2 + 2b_1 = 6a_2x_2 + 2b_2
\end{cases}
\end{equation}

```python
x = [0.1, 0.5, 0.9, 1.3, 1.7]
y = [-2.3026, -0.69315, -0.10536, 0.26236, 0.53063]

A = np.array([
    [x[0]**3, x[0]**2, x[0], 1, 0, 0, 0, 0],
    [x[1]**3, x[1]**2, x[1], 1, 0, 0, 0, 0],
    [x[2]**3, x[2]**2, x[2], 1, 0, 0, 0, 0],
    [0, 0, 0, 0, x[2]**3, x[2]**2, x[2], 1],
    [0, 0, 0, 0, x[3]**3, x[3]**2, x[3], 1],
    [0, 0, 0, 0, x[4]**3, x[4]**2, x[4], 1],
    [3*(x[2]**2), 2*x[2], 1, 0, -3*(x[2]**2), -2*x[2], -1, 0],
    [6*x[2], 2, 0, 0, -6*x[2], -2, 0, 0]
])
f = np.array([y[0], y[1], y[2], y[2], y[3], y[4], 0, 0])

a1,b1,c1,d1,a2,b2,c2,d2 = np.linalg.solve(A,f)
```

```python
def polynom(x, coefs):
    if x == 0:
        return coefs[0]
    res = 0
    i = 0
    for c in coefs:
        res += coefs[i]*(x**i)
        i += 1
    return res

def spline(x, net, coefs):
    if x < net[2]:
        return polynom(x, coefs[:4])
    else:
        return polynom(x, coefs[4:])

x_test = np.linspace(min(x), max(x), 100)
y_test = []
for p_test in x_test:
    y_test.append(spline(p_test, x, [d1, c1, b1, a1, d2, c2, b2, a2]))
plt.scatter(x, y)
plt.plot(x_test, y_test)
plt.grid()
plt.show()
```

$I \approx \int\limits_a^b \text{spline}(x)\sin(x) dx = \int\limits_{x_0}^{x_2} (a_1x^3 + b_1x^2 + c_1x + d_1)\sin(x)dx + \int\limits_{x_2}^{x_5} (a_2x^3 + b_2x^2 + c_2x + d_2)\sin(x)dx$. Далее беря интегралы от $x^p\sin(x)$ и умножая на коэффициенты получаем:

```python
np.trapz(y_test, x_test, dx = 0.001)
```

Итого, ответ $I \approx  -0.494$
